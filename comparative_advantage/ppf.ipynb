### Create Figures for Comparative Advantage Introduction

```python
import numpy as np
import matplotlib.pyplot as plt
```

Create agent dictionary with production capabilities.

```python
agent_dict = {
    'Agent 1': {
        'prod_g1': 60,
        'prod_g2': 20
    },
    'Agent 2': {
        'prod_g1': 100,
        'prod_g2': 80
    }
}
```

Plot PFF with autarky production.

```python
# Create figure and axis. Good 1 is the y-axis and Good 2 is the x-axis.
fig, ax = plt.subplots()

for i, a in enumerate(agent_dict.keys()):
    color = next(ax._get_lines.prop_cycler)['color']

    # Plot Agent's PPF.
    x = np.linspace(0, agent_dict[a]['prod_g2'])
    m = (agent_dict[a]['prod_g1'] - 0) / (0 - agent_dict[a]['prod_g2'])
    b = agent_dict[a]['prod_g1']
    y = m*x + b

    plt.plot(x, y, linestyle='-', label=f"Agent {i+1}", color=color)

    # Plot Agent's autarky production.
    aut = (agent_dict[a]['prod_g2'] / 2, agent_dict[a]['prod_g1'] / 2)
    plt.plot(aut[0], aut[1], marker='o', color=color)
    ax.text(aut[0], aut[1]+3, f"({str(round(aut[0], 0))}, {str(round(aut[1], 0))})", fontsize=12)

# Set graph title, axis labels, axis window, and legend.
plt.title('Production Possibility Frontier')

plt.xlabel('Quantity of Good 1')
plt.ylabel('Quantity of Good 2')

ax.set_ylim(0, max([agent_dict[a]['prod_g1'] for a in agent_dict.keys()]))
ax.set_xlim(0, max([agent_dict[a]['prod_g2'] for a in agent_dict.keys()]))

plt.legend()

# Output graph.
plt.savefig('static/ppf.png')
```
